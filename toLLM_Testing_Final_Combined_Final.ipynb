### AUTOMATE PROCESS AND SAVE RESULTS

import json
import ollama
import chromadb
import os
from datetime import datetime

# Fetch entries from ChromaDB, including metadata and documents.
def fetch_entries_from_chromadb(collection_name="news_articles", max_entries=5):
    """ Fetches the latest enriched entries from ChromaDB. """
    try:
        # Initialize the ChromaDB client.
        client = chromadb.PersistentClient(path="./chroma_db")

        # Get the specified collection from ChromaDB.
        collection = client.get_collection(name=collection_name)
        # Retrieve documents with metadata, limiting to 'max_entries' items.
        results = collection.get(include=["metadatas", "documents"], limit=max_entries)

        entries = []
        # Loop through the documents to extract the necessary data.
        for i in range(len(results["documents"])):
            entry = {
                # Fetch metadata information or set defaults if not available.
                "title": results["metadatas"][i].get("title", "Unknown Title"),
                "url": results["metadatas"][i].get("url", "Unknown URL"),
                "published_date": results["metadatas"][i].get("published_date", "Unknown Date"),
                "source": results["metadatas"][i].get("source", "Unknown Source"),
                "author": results["metadatas"][i].get("author", "Unknown Author"),
                # Store the enriched content (document itself).
                "enriched_content": results["documents"][i]
            }
            entries.append(entry)
        return entries
    except Exception as e:
        # Catch errors during the fetching process and print the error message.
        print(f"Error fetching data from ChromaDB: {e}")
        return []

# Construct the prompt for fact-checking based on the article.
def construct_fact_checking_prompt(entry):
    """ Constructs a fact-checking prompt for the LLM. """
    return f"""
    Analyze the following news article:

    Title: {entry['title']}
    Source: {entry['source']}
    Published: {entry['published_date']}
    Content: {entry['enriched_content']}

    Is this news article reliable? Provide a fact-checking analysis.
    """

# Generate a unique filename for saving results in the file system.
def get_unique_filename(base_path="PROMPT_RESULTS"):
    """ Generates a unique filename for storing responses. """
    # Create the directory if it doesn't already exist.
    os.makedirs(base_path, exist_ok=True)
    # Format the current date as a string.
    date_str = datetime.now().strftime("%Y-%m-%d")
    file_number = 1

    # Continuously check for a unique file number to avoid overwriting files.
    while True:
        filename = f"Prompt_Return_{date_str}_{file_number}.json"
        full_path = os.path.join(base_path, filename)
        if not os.path.exists(full_path):
            return full_path
        file_number += 1

# Analyze multiple articles using specified LLM models.
def analyze_multiple_articles(chroma_entries, models=["deepseek-r1", "llama3"]):
    """
    Sends multiple articles to specified LLM models for analysis and ensures results are structured per model.
    """
    responses = []

    # Iterate through each article entry retrieved from ChromaDB.
    for entry in chroma_entries:
        # Construct the fact-checking prompt for the current article.
        prompt_text = construct_fact_checking_prompt(entry)
        fact_check_results = {}

        # Iterate through the list of models and send the prompt to each.
        for model in models:
            try:
                # Send the prompt to the LLM model and capture the response.
                response = ollama.chat(
                    model=model,
                    messages=[{"role": "user", "content": prompt_text}]
                )
                # Store the response content for each model.
                fact_check_results[model] = response['message']['content']
            except Exception as e:
                # In case of error with the LLM request, store an error message.
                fact_check_results[model] = f"Error processing request: {e}"

        # Append the analysis results for the current article.
        responses.append({
            # Include metadata such as title, URL, and source.
            "title": entry.get("title", "Unknown Title"),
            "url": entry.get("url", "Unknown URL"),
            "published_date": entry.get("published_date", "Unknown Date"),
            "source": entry.get("source", "Unknown Source"),
            "author": entry.get("author", "Unknown Author"),
            # Store fact-checking results, including responses from all models.
            "fact_check_results": fact_check_results  # Store the results per model.
        })

    # Save the results to a unique JSON file.
    output_file = get_unique_filename()
    try:
        # Open the file in write mode with UTF-8 encoding.
        with open(output_file, "w", encoding="utf-8") as file:
            # Dump the responses to the JSON file, formatted for readability.
            json.dump(responses, file, indent=4, ensure_ascii=False)
        print(f"Results saved to: {output_file}")
    except Exception as e:
        # Handle any errors during the file saving process.
        print(f"Error saving results: {e}")

    return responses

# ** Execute the analysis **
# Fetch the most recent entries from ChromaDB.
chroma_entries = fetch_entries_from_chromadb(max_entries=5)
if chroma_entries:
    # If data is retrieved, proceed to analyze the articles.
    results = analyze_multiple_articles(chroma_entries)
    # Print the results in a formatted JSON style.
    print(json.dumps(results, indent=4, ensure_ascii=False))
else:
    # If no data is retrieved, print a message.
    print("No data retrieved from ChromaDB.")

```python
import json
import ollama
import chromadb
import os
from datetime import datetime

# Fetch entries from ChromaDB, including metadata and documents.
def fetch_entries_from_chromadb(collection_name="news_articles", max_entries=5):
    """ Fetches the latest enriched entries from ChromaDB. """
    try:
        # Initialize the ChromaDB client.
        client = chromadb.PersistentClient(path="./chroma_db")

        # Get the specified collection from ChromaDB.
        collection = client.get_collection(name=collection_name)
        # Retrieve documents with metadata, limiting to 'max_entries' items.
        results = collection.get(include=["metadatas", "documents"], limit=max_entries)

        entries = []
        # Loop through the documents to extract the necessary data.
        for i in range(len(results["documents"])):
            entry = {
                # Fetch metadata information or set defaults if not available.
                "title": results["metadatas"][i].get("title", "Unknown Title"),
                "url": results["metadatas"][i].get("url", "Unknown URL"),
                "published_date": results["metadatas"][i].get("published_date", "Unknown Date"),
                "source": results["metadatas"][i].get("source", "Unknown Source"),
                "author": results["metadatas"][i].get("author", "Unknown Author"),
                # Store the enriched content (document itself).
                "enriched_content": results["documents"][i]
            }
            entries.append(entry)
        return entries
    except Exception as e:
        # Catch errors during the fetching process and print the error message.
        print(f"Error fetching data from ChromaDB: {e}")
        return []

# Construct the prompt for fact-checking based on the article.
def construct_fact_checking_prompt(entry):
    """ Constructs a fact-checking prompt for the LLM. """
    return f"""
    Analyze the following news article:

    Title: {entry['title']}
    Source: {entry['source']}
    Published: {entry['published_date']}
    Content: {entry['enriched_content']}

    Is this news article reliable? Provide a fact-checking analysis.
    """

# Generate a unique filename for saving results in the file system.
def get_unique_filename(base_path="PROMPT_RESULTS"):
    """ Generates a unique filename for storing responses. """
    # Create the directory if it doesn't already exist.
    os.makedirs(base_path, exist_ok=True)
    # Format the current date as a string.
    date_str = datetime.now().strftime("%Y-%m-%d")
    file_number = 1

    # Continuously check for a unique file number to avoid overwriting files.
    while True:
        filename = f"Prompt_Return_{date_str}_{file_number}.json"
        full_path = os.path.join(base_path, filename)
        if not os.path.exists(full_path):
            return full_path
        file_number += 1

# Analyze multiple articles using specified LLM models.
def analyze_multiple_articles(chroma_entries, models=["deepseek-r1", "llama3"]):
    """
    Sends multiple articles to specified LLM models for analysis and ensures results are structured per model.
    """
    responses = []

        # Iterate through each article entry retrieved from ChromaDB.
    for entry in chroma_entries:
        # Construct the fact-checking prompt for the current article.
        prompt_text = construct_fact_checking_prompt(entry)
        fact_check_results = {}

        # Iterate through the list of models and send the prompt to each.
        for model in models:
            try:
                # Send the prompt to the LLM model and capture the response.
                print(f"Sending request to {model}...")  # Add logging to ensure both models are called
                response = ollama.chat(
                    model=model,
                    messages=[{"role": "user", "content": prompt_text}]
                )
                # Store the response content for each model.
                fact_check_results[model] = response['message']['content']
            except Exception as e:
                # In case of error with the LLM request, store an error message.
                print(f"Error processing {model}: {e}")
                fact_check_results[model] = f"Error processing request: {e}"

        # Append the analysis results for the current article.
        responses.append({
            # Include metadata such as title, URL, and source.
            "title": entry.get("title", "Unknown Title"),
            "url": entry.get("url", "Unknown URL"),
            "published_date": entry.get("published_date", "Unknown Date"),
            "source": entry.get("source", "Unknown Source"),
            "author": entry.get("author", "Unknown Author"),
            # Store fact-checking results, including responses from all models.
            "fact_check_results": fact_check_results  # Store the results per model.
        })

    # Save the results to a unique JSON file.
    output_file = get_unique_filename()
    try:
        # Open the file in write mode with UTF-8 encoding.
        with open(output_file, "w", encoding="utf-8") as file:
            # Dump the responses to the JSON file, formatted for readability.
            json.dump(responses, file, indent=4, ensure_ascii=False)
        print(f"Results saved to: {output_file}")
    except Exception as e:
        # Handle any errors during the file saving process.
        print(f"Error saving results: {e}")


    return responses

# ** Execute the analysis **
# Fetch the most recent entries from ChromaDB.
chroma_entries = fetch_entries_from_chromadb(max_entries=5)
if chroma_entries:
    # If data is retrieved, proceed to analyze the articles.
    results = analyze_multiple_articles(chroma_entries)
    # Print the results in a formatted JSON style.
    print(json.dumps(results, indent=4, ensure_ascii=False))
else:
    # If no data is retrieved, print a message.
    print("No data retrieved from ChromaDB.")
```

```python
import pandas as pd
import json

# Assuming 'results' is the list containing news articles with fact-checking results
print(f"Number of articles in results: {len(results)}")
print("Example of an article in results:")
print(results[0])

# Extract relevant information and create a labeled DataFrame
labeled_data = []

for article in results:
    title = article.get("title", "")
    url = article.get("url", "")
    published_date = article.get("published_date", "")
    source = article.get("source", "")
    author = article.get("author", "")
    fact_check_results = article.get("fact_check_results", {})
    
    # Extract evaluation text from fact-checking results
    deepseek_result = fact_check_results.get("deepseek-r1", "")
    llama_result = fact_check_results.get("llama3", "")
    
    # Define a simple reliability scoring function
    def compute_reliability_score(deepseek_text, llama_text):
        keywords_positive = ["reliable", "credible", "trustworthy", "high confidence", "accurate"]
        keywords_negative = ["unreliable", "uncertain", "low confidence", "doubtful", "untrustworthy"]
        
        score = 0.5  # Start with a neutral score
        
        for keyword in keywords_positive:
            if keyword in deepseek_text.lower() or keyword in llama_text.lower():
                score += 0.125  # Increase reliability
        
        for keyword in keywords_negative:
            if keyword in deepseek_text.lower() or keyword in llama_text.lower():
                score -= 0.125  # Decrease reliability
        
        return max(0, min(1, score))  # Ensure score remains in [0,1]
    
    reliability_score = compute_reliability_score(deepseek_result, llama_result)
    
    labeled_data.append({
        "title": title,
        # "url": url,
        "published_date": published_date,
        "source": source,
        "author": author,
        "fact_check_results": fact_check_results,
        "reliability_score": reliability_score
    })

# Convert labeled data to DataFrame
labeled_df = pd.DataFrame(labeled_data)

# Save labeled data to CSV
# labeled_df.to_csv("labeled_results.csv", index=False)
# print("Labeled results saved to CSV.")

# Display the first few rows
labeled_df.head()
```
